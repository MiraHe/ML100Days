### 匯入所需套件

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### 輸入資料

```python
# 輸入資料
file_path = r"D:\Github\ML100Days\train.csv"

df = pd.read_csv(file_path)
df.head()
```

### 繪製長條圖

```python
# 收入分佈是否偏態？是否存在極端高收入？→ 用直方圖 (Histogram) 查看 income 分佈
plt.figure(figsize=(8,4))
plt.hist(df['annual_income'].dropna(), bins=50)
plt.title("Annual Income Distribution")
plt.xlabel("annual_income")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

### 繪製盒鬚圖

```python
# 負債比是否有極端值？可能影響模型？→ 盒鬚圖 (Boxplot) 可偵測離群點
plt.figure(figsize=(6,4))
plt.boxplot(df['debt_to_income_ratio'].dropna())
plt.title("Debt to Income Ratio - Boxplot (Detect Outliers)")
plt.ylabel("debt_to_income_ratio")
plt.tight_layout()
plt.show()
```

### 繪製散佈圖

```python
# 收入越高，信用分數是否越高？→ 散佈圖 (Scatter Plot) 觀察變數相關性
plt.figure(figsize=(6,4))
plt.scatter(df['annual_income'], df['credit_score'], alpha=0.5)
plt.title("Annual Income vs Credit Score (Trend Insight)")
plt.xlabel("annual_income")
plt.ylabel("credit_score")
plt.tight_layout()
plt.show()
```

### 繪製折線圖

```python
# 貸款金額是否依資料順序呈現週期/規律？→ 折線圖 (Line Plot) 看排序後貸款趨勢是否有規律
loan_sorted = df['loan_amount'].sort_values().reset_index(drop=True)

plt.figure(figsize=(8,4))
plt.plot(loan_sorted)
plt.title("Loan Amount Trend (Sorted to Reveal Pattern)")
plt.xlabel("Sorted Index")
plt.ylabel("loan_amount")
plt.tight_layout()
plt.show()
```
