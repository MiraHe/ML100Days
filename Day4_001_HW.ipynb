## Part2. 請上「政府資料開放平臺」，找到感興趣的資料集（限定檔案格式要 CSV 檔或是 Excel 檔），並匯入至 Python，進一步使用 .loc、.iloc 挑選出特定的資料

```python
""" 請匯入 CSV 檔案或是 Excel 檔案 """
import pandas as pd

# 路徑
file_path = r"D:\Github\ML100Days\train.csv"
#讀取
df = pd.read_csv(file_path)
df.head(5)
```

```python
""" 請使用 .loc 挑選出三組感興趣的資料，並詳細說明對哪三組資料感興趣"""
# 新增一個欄位「income_to_loan_ratio」 = 年收入 / 貸款金額
# 如果這個比值越高，代表收入相對於貸款金額越充足，還款能力強。
# 如果這個比值越低，代表借得多但收入低，風險高。
df["income_to_loan_ratio"] = df["annual_income"] / df["loan_amount"]

# 找出高收入且利率高，但仍成功還款的前 3 筆
interesting_loc = df.loc[
    (df["annual_income"] > df["annual_income"].mean()) &
    (df["interest_rate"] > df["interest_rate"].mean()) &
    (df["loan_paid_back"] == 1)
].sort_values(by="interest_rate", ascending=False).head(3)

interesting_loc
```

```python
""" 請使用 .iloc 挑選出三組感興趣的資料，並詳細說明對哪三組資料感興趣"""
# 按信用分數從低到高排序
data_sorted = df.sort_values(by="credit_score", ascending=True)

# 用 iloc 選出前 3 筆（最低信用分數者）
interesting_iloc = data_sorted.iloc[0:3]

interesting_iloc

#1️⃣ 這三筆資料代表「信用評分最低」的借款人，→ 可以分析這些人是否確實無法還款（loan_paid_back == 0），→ 以驗證信用分數是否真能反映違約風險。
#2️⃣ 若這些人中仍有人成功還款，可能說明信用評分模型需調整。
#3️⃣ 可觀察他們的收入、負債比、職業是否偏離一般違約族群。
```
