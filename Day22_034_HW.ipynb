可以看見 y 類別中，有 160 個 類別 0，40 個 類別 1 ，請試著使用 train_test_split 函數，切分出 y_test 中能各有 10 筆類別 0 與 10 筆類別 1 。(HINT: 參考函數中的 test_size，可針對不同類別各自作切分後再合併)

```python
from sklearn.model_selection import train_test_split
```

```python
X = np.arange(1000).reshape(200, 5)
y = np.zeros(200, dtype=int)
y[:40] = 1     # 40 個 1、160 個 0

# 1. 先把 0 類別與 1 類別分開
X_1 = X[y == 1]
y_1 = y[y == 1]

X_0 = X[y == 0]
y_0 = y[y == 0]

# 2. 各自用 train_test_split 取出 10 筆當 test
X1_train, X1_test, y1_train, y1_test = train_test_split(
    X_1, y_1, test_size=10, random_state=42)

X0_train, X0_test, y0_train, y0_test = train_test_split(
    X_0, y_0, test_size=10, random_state=42)

# 3. 再把兩類別的 train / test 合併
X_train = np.vstack([X1_train, X0_train])
y_train = np.concatenate([y1_train, y0_train])

X_test = np.vstack([X1_test, X0_test])
y_test = np.concatenate([y1_test, y0_test])
```
